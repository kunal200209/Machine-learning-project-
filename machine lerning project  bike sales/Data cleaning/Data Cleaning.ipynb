```python
# Identifying Unqiue values for each column 
for i in df:
    print(f"{i} : {df[i].unique()}")
    print("***************************")
```

```python
#Indentifying nan values for each colum,n so that we can change them
for i in df:
    print(f"{i} : {df[i].isna().sum()}")
    print("***************************")
```

```python
# Step 1: Convert the 'Date' column to datetime format if it isn't already
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Step 2: Fill missing values in 'Day' column with the derived day
df['Day'] = df['Day'].fillna(df['Date'].dt.day)


# Now, the 'Day' column should be filled with the correct values based on the 'Date' column.
print(df[['Date', 'Day']].head())
```

```python
#Age_Group : ['Adults (35-64)' 'Young Adults (25-34)' 'Youth (<25)' nan]
for i in df['Customer_Age']:
    if i < 25:
        df['Age_Group']=df['Age_Group'].fillna('Youth (<25)')
    elif i >= 25  and i<=34:
        df['Age_Group']=df['Age_Group'].fillna('Young Adults (25-34)')
    elif i >= 35  and i<=64:
        df['Age_Group']=df['Age_Group'].fillna('Young Adults (25-34)')
```

Some of the columns like unit cost, revenue have 0 in them So this rows should be dropped

```python

# Replacing  0 with NaN in the entire DataFrame so that we can drop the whole record 
df = df.replace(0, np.nan)
for i in df:
    print(f"{i} : {df[i].isna().sum()}")
    print("***************************")
```

Dropping the records which have nan values

```python
# Drop rows where any column has NaN values
data_cleaned = df.dropna(inplace=True)



for i in df:
    print(f"{i} : {df[i].isna().sum()}")
    print("***************************")

```

##Converting the categorical columns into numerical comlumns beacuse ml algorithm oly works on numerical values

```python

from sklearn.preprocessing import LabelEncoder

cf=df[['Month','Age_Group','Customer_Gender','Country','State','Product_Category','Sub_Category','Product_Description']]

# Initialize the LabelEncoder
le = LabelEncoder()



# Apply label encoding to each categorical column
for i in cf:
    df[i] = le.fit_transform(df[i].astype(str))

# Check the first few rows to confirm label encoding
df
```
