```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = r'C:\Users\User\Desktop\machine lerning project  bike sales\Data cleaning\Clean_data.xlsx'

df = pd.read_excel(file_path)

df.columns = df.columns.str.strip()
print(df)
print(df.isna().sum())
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Profit')

plt.xlabel('Date')
plt.ylabel('Profit')
plt.title('Profit by Date')
plt.grid(True)

plt.show()
```

```python
df['Age_Group'].value_counts().plot(kind = 'barh', figsize = (10,2), grid = True, color='teal')
plt.title('Age Groups Amount')
plt.xlabel('Amount')
plt.ylabel('Age Group')
plt.show()
```

```python
df['Product_Description'].value_counts(ascending = False).head().plot(kind = 'barh', figsize = (10,2), grid = True, color='teal')
plt.title('Product Description Amount')
plt.xlabel('Amount')
plt.ylabel('Product Description')
plt.show()
```

```python
for col in df:
    if col == 'Sales_Order #':
        pass
    else:
        sns.countplot(data=df,y=col,palette="deep",order=df[col].value_counts().index)
        plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
df = df.drop(['Product_Category', 'Sub_Category'], axis=1)


# Step 1: Check for missing values and handle them (if necessary)
# Optionally, fill NaN values for numerical columns (if required)
# df.fillna(0, inplace=True)

# Step 2: Calculate the correlation matrix for numerical columns
corr_matrix = df.corr()

# Step 3: Plot the heatmap
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Step 4: Display the heatmap
plt.title('Correlation Heatmap')
plt.show()


```
